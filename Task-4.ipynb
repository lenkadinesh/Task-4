# Kaggle Raw Data

# Data set loading 

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\ocdga\Downloads\archive\kaggle-raw.csv")
print(df)
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```
